```python
import requests
from bs4 import BeautifulSoup

# ... headers, id_questionario, lista_questoes já definidos ...

html_parts = []
html_parts.append(f"""<!DOCTYPE html>
<html lang="pt-BR">
<head>
  <meta charset="utf-8">
  <title>Questionário {id_questionario}</title>
  <meta name="viewport" content="width=device-width, initial-scale=1">
  <style>
    body {{ font-family: system-ui, -apple-system, Segoe UI, Roboto, Arial, sans-serif; margin: 24px; }}
    .container {{ max-width: 900px; margin: 0 auto; }}
    .meta {{ color:#444; margin-bottom: 16px; }}
    fieldset {{ border: 1px solid #ddd; padding: 16px 20px; margin: 16px 0; border-radius: 12px; }}
    legend {{ padding: 0 8px; font-weight: 600; }}
    .alternativa {{ display: flex; align-items: flex-start; gap: 8px; margin: 8px 0; }}
    .alternativa input {{ margin-top: 3px; }}
    .enunciado {{ font-size: 1.05rem; margin-bottom: 8px; }}
    .q-idx {{ color: #666; font-size: .9rem; margin-left: 6px; }}
    button[type="submit"] {{ padding: 10px 16px; border-radius: 10px; border: 1px solid #bbb; background: #f6f6f6; cursor: pointer; }}
    button[type="submit"]:hover {{ background: #eee; }}
  </style>
</head>
<body>
  <div class="container">
    <h1>Questionário</h1>
    <div class="meta"><strong>ID do questionário:</strong> {id_questionario}</div>
    <form>
""")

for idx, questao in enumerate(lista_questoes, start=1):
    id_questao = questao['idQuestionarioAlunoItem']

    # Enunciado
    url_enunciado = (
        f"https://studeoapi.unicesumar.edu.br/avaliacao-api-controller/api/"
        f"questionario/info-questao/questionario-aluno/{id_questionario}/"
        f"questionario-aluno-item/{id_questao}"
    )
    enunciado = requests.get(url=url_enunciado, headers=headers).json().get('dsEnunciado', '')

    # Alternativas
    url_alternativas = (
        f"https://studeoapi.unicesumar.edu.br/avaliacao-api-controller/api/"
        f"questionario/info-alternativa/questionario-aluno/{id_questionario}/"
        f"questionario-aluno-item/{id_questao}"
    )
    lista_alternativas_json = requests.get(url=url_alternativas, headers=headers).json()

    html_parts.append('      <fieldset>')
    html_parts.append(f'        <legend>Questão {idx} <span class="q-idx">ID {id_questao}</span></legend>')
    html_parts.append(f'        <div class="enunciado">{enunciado}</div>')

    group_name = f"q_{id_questao}"

    for alternativa in lista_alternativas_json:
        alt_id = alternativa.get('idQuestionarioAlunoItemAlternativa')
        alt_texto = alternativa.get('dsAlternativa', '')
        checked_attr = ' checked' if alternativa.get('flSelecionado') else ''
        input_id = f"alt_{alt_id}"

        # Remove tags <p> mantendo conteúdo
        soup = BeautifulSoup(alt_texto, "html.parser")
        # se a alternativa inteira está dentro de <p>, pegar só o conteúdo interno
        if soup.p and len(soup.contents) == 1:
            alt_texto_limpo = soup.p.decode_contents()
        else:
            alt_texto_limpo = soup.decode_contents()

        html_parts.append(
            f'        <label class="alternativa" for="{input_id}">'
            f'          <input type="radio" id="{input_id}" name="{group_name}" value="{alt_id}"{checked_attr}>'
            f'          <span>{alt_id} - {alt_texto_limpo}</span>'
            f'        </label>'
        )

    html_parts.append('      </fieldset>')

html_parts.append("""      <button type="submit">Enviar</button>
    </form>
  </div>
</body>
</html>
""")

arquivo = "prova.html"
with open(arquivo, "w", encoding="utf-8") as f:
    f.write("\n".join(html_parts))

print(f"Arquivo gerado: {arquivo}")
```
