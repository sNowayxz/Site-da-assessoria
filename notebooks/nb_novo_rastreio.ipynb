```python
import html
import re

def normalizar_descricao_texto(texto):
    # 1. Descodifica entidades HTML como &eacute;, &ccedil; etc.
    texto = html.unescape(texto)

    # 2. Remove espaços HTML extras (&nbsp;) e converte para espaço comum
    # texto = texto.replace('\xa0', ' ').replace('&nbsp;', ' ')

    # 3. Substitui quebras de linha HTML (\r\n ou \n) por parágrafos reais
    texto = re.sub(r'(\r\n|\n)+', '\n\n', texto)

    # 4. Remove espaços extras múltiplos
    texto = re.sub(r'[ ]{2,}', ' ', texto)

    # 5. Tira espaços excessivos no início e fim
    return texto.strip()

# Exemplo de uso:
descricao_bruta = dados[0]['descricaoTexto']
descricao_limpa = normalizar_descricao_texto(descricao_bruta)
print(descricao_limpa)
hash_text(descricao_limpa)
```

```python
from bs4 import BeautifulSoup
import html

def extrair_texto_com_quebras(html_input):
    soup = BeautifulSoup(html_input, "html.parser")
    partes = []

    for elem in soup.recursiveChildGenerator():
        if elem.name == 'br':
            partes.append('\n\n')
        elif elem.name is None:
            texto = html.unescape(elem.strip())
            if texto:
                partes.append(texto)

    texto_final = ''.join(partes)
    # Remove múltiplas quebras e espaços extras
    return '\n\n'.join([par.strip() for par in texto_final.split('\n\n') if par.strip()])


entrada_html = """<div class="enunciado ng-binding" ng-bind-html="vm.questaoAtual.descricaoHtml  | mathJaxFix"><div class="ng-scope">Certamente, você já deve ter escutado que o que destaca o ser humano dos demais&nbsp;animais é a sua vasta capacidade de adaptação aos ambientes mais variados. Ele é&nbsp;capaz de habitar em florestas, cidades, enfrentar o frio ou suportar o calor. Ele modifica o&nbsp;ambiente ao seu redor para garantir sua sobrevivência com conforto. Isso é possível&nbsp;graças a uma habilidade extraordinária: a de aprender.</div>
<br class="ng-scope">
<br class="ng-scope">
Considerando o dado fragmento de texto e considerando a capacidade simbólica de&nbsp;operação do ser humano, assinale a alternativa que melhor descreve essa habilidade:<br class="ng-scope">
<br class="ng-scope">
&nbsp;</div>"""

texto_extraido = extrair_texto_com_quebras(entrada_html)
print(texto_extraido)
```

```python
def gerar_descricao_texto(texto_limpo: str) -> str:
    paragrafos = texto_limpo.strip().split('\n\n')
    paragrafos_escapados = [html.escape(par.strip()) for par in paragrafos if par.strip()]
    return '\r\n\r\n'.join(paragrafos_escapados) + '\r\n\r\n&nbsp;'
```
