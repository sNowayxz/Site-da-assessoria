```python
import html
from bs4 import BeautifulSoup
import re

html_input = '''<div class="enunciado ng-binding" ng-bind-html="vm.questaoAtual.descricaoHtml  | mathJaxFix"><div class="ng-scope">Para obtermos uma imagem radiográfica com qualidade, é necessário a junção de três princípios básicos: posicionamento, técnica e colimação.<br/> No processo da formação da imagem, esses princípios se complementam.<br/> O posicionamento é considerado o mais importante dentre os três. Estima-se que, em 85% de todos os exames que necessitam de repetição, a causa seja por erro no posicionamento. Além de ser fundamental na identificação de possíveis alterações na imagem, como na identificação de uma pequena fratura, por exemplo.<br/> Outro princípio, a respeito da técnica, ou mais precisamente, aos fatores de exposição, Kvp e mAs selecionados, sendo o Kvp responsável pelo contraste e pela qualidade de penetração da imagem, é o mAs, responsável pela densidade ou brilho na imagem, bem como o ruído.<br/> Com relação ao terceiro pilar, a colimação, percebemos que muitas vezes é negligenciado pelos profissionais, ainda mais na geração atual dos exames digitais.<br/> <br/> Fonte:  BREY, A. C. <strong>Técnicas radiológicas de membros superiores e inferiores</strong>. Florianópolis: Arqué, 2024.<br/> <br/> Considerando o princípio da colimação:<br/> a) <strong>DESCREVA</strong> o que é e como é utilizada na realização de um exame radiográfico, bem como sua importância na qualidade da imagem.<br/> b) <strong>APRESENTE</strong> qual a relação da colimação com a proteção radiológica no ambiente.</div></div>'''

from html.entities import codepoint2name

# 1. Extrair texto do HTML (com quebras de linha)
soup = BeautifulSoup(html_input, 'html.parser')
conteudo = soup.get_text(separator='\r\n')

# 2. Função que converte todos os caracteres especiais para entidades nomeadas, se existir
def to_named_html_entities(text):
    def char_to_entity(c):
        codepoint = ord(c)
        if codepoint in codepoint2name:
            return f"&{codepoint2name[codepoint]};"
        else:
            return c
    return ''.join(char_to_entity(c) for c in text)

conteudo_final = to_named_html_entities(conteudo)

conteudo_final = re.sub(r'(\r\n)[ ]+', r'\1', conteudo_final)

# 3. Visualização de um trecho para conferir (até 300 chars)
conteudo_final
```
