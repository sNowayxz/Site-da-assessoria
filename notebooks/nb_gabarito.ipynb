```python
import requests as rq

def hash_text(text)->str:
    import hashlib

    # Cria um objeto hash com SHA-256
    hash_object = hashlib.sha256()
    
    # Codifica o texto em UTF-8 e atualiza o objeto hash
    hash_object.update(text.encode('utf-8'))
    
    # Obtém o hash hexadecimal
    hex_dig = hash_object.hexdigest()
    
    return hex_dig
```

```python
lista_alternativas = [
    'As asserções I e II são proposições verdadeiras, e a II é uma justificativa correta da I.',
    'As asserções I e II são proposições verdadeiras, mas a II não é uma justificativa correta da I.',
    'A asserção I é uma proposição verdadeira, e a II é uma proposição falsa. ',
    'A asserção I é uma proposição falsa, e a II é uma proposição verdadeira. ',
    'As asserções I e II são proposições falsas. '
]

url_base ="https://www.papiron.com.br"

enunciado = """<div class="enunciado ng-binding" ng-bind-html="vm.questaoAtual.descricaoHtml  | mathJaxFix"><p class="ng-scope">Ao longo da história, o conceito de criatividade passou por diversas transformações. Segundo De Bono (1992), a criatividade era vista como um dom inato de alguns indivíduos, uma habilidade mágica que surgia de forma espontânea e que não podia ser ensinada. Porém, com o tempo, esta visão foi sendo modificada e, atualmente, acredita-se que a criatividade é uma habilidade que pode ser desenvolvida e estimulada em qualquer pessoa, para isso, é fundamental contar com um ambiente propício à inovação (NAKAGAWA, 2014). A criatividade também tem um papel fundamental na resolução de problemas e na tomada de decisões.</p> <p class="ng-scope"> Fonte: adaptada de: ZUCCO, A. et al.<strong> Mentalidade Criativa e Empreendedora.</strong> Indaial, SC: Arqué, 2023.</p> <p class="ng-scope">Em relação à criatividade, observe as asserções a seguir e a relação proposta entre elas:</p> <p class="ng-scope">I. Em qualquer negócio a incerteza é algo fácil de ser encontrado. Como aliada, a criatividade pode ser uma excelente parceira.</p> <p class="ng-scope">PORQUE</p> <p class="ng-scope">II. O empreendedor criativo é capaz de encontrar inovadoras soluções e consequentemente ser capaz de tomar decisões mais acertadas.</p> <p class="ng-scope">A respeito dessas asserções, assinale a opção correta.</p></div>"""

hash_enunciado = hash_text(enunciado)

gabarito = ""

url_hash = url_base+"/atividades/gabarito_hash/"+hash_enunciado

r = rq.get(url_hash, allow_redirects=False)

data_enunciados = r.json()

data_enunciados

print("\n           ", 
    list(data_enunciados['0']['alternativas'].values()) == lista_alternativas,
    "\n       >> "+str(list(data_enunciados['0']['alternativas'].values()))+"\n      >> "+str(lista_alternativas))
```

```python
# obtém as alternativas do dict data_enunciados
alternativas_dict = data_enunciados['0']['alternativas'].values()

# verifica se todas as alternativas da lista estão no dict
todas_presentes = all(texto in alternativas_dict for texto in lista_alternativas)

todas_presentes

if todas_presentes:
    gabarito = data_enunciados['0']['gabarito']
else:
    gabarito = None

gabarito
```
