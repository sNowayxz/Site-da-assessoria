```python
from bs4 import BeautifulSoup
import json

# Lê o arquivo HTML
with open('CÁLCULO I_5200997.html', encoding='utf-8') as f:
    html = f.read()

soup = BeautifulSoup(html, "html.parser")
container = soup.find("div", class_="avaliacoes")

questoes = []
questao_atual = None

for node in container.find_all(recursive=False):
    if "questao" in node.get("class", []):
        titulo = node.find("div", class_="titulo").get_text(" ", strip=True)
        enunciado_div = node.find("div", class_="enunciado")
        if enunciado_div:
            enunciado = enunciado_div.decode_contents().strip()
        else:
            enunciado = ""
        questao_atual = {
            "titulo": titulo,
            "enunciado": enunciado,
            "gabarito": None,
            "alternativas": []
        }
        questoes.append(questao_atual)
    elif "alternativa" in node.get("class", []) and questao_atual:
        letra = node.find("div", class_="ordem").get_text(" ", strip=True)
        enunciado_alt = node.find("div", class_="enunciado")
        texto = enunciado_alt.get_text(" ", strip=True) if enunciado_alt else ""

        # É a alternativa correta do gabarito? (pela classe "correta", não pelo X)
        correta = (
            "correta" in node.get("class", []) or
            (node.find("div", class_="ordem") and "correta" in node.find("div", class_="ordem").get("class", [])) or
            (node.find("div", class_="enunciado") and "correta" in node.find("div", class_="enunciado").get("class", []))
        )
        if correta:
            questao_atual["gabarito"] = texto

        # Lista de alternativas (apenas o texto, ou se quiser letra junto: f"{letra}) {texto}")
        questao_atual["alternativas"].append(texto)

# Remove o campo gabarito None se não houver (opcional)
for q in questoes:
    if q["gabarito"] is None:
        del q["gabarito"]

# Salva em JSON
with open('questoes_json_gabarito.json', 'w', encoding='utf-8') as out:
    json.dump(questoes, out, ensure_ascii=False, indent=2)

# Exemplo de impressão:
print(json.dumps(questoes, ensure_ascii=False, indent=2))
```

```python

def listar_templates(itens):
    import re


    # Palavras que DEVEM aparecer na descrição (case insensitive)
    palavras_incluir = [
        "MAPA", "TEMPLATE", "FORMULÁRIO", "FORMULARIO",
        "MODELO", "PADRÃO", "AE1", "ATIVIDADE 1",
        "RELATÓRIO", "ERRATA"
    ]
    # Palavras que NÃO PODEM aparecer na descrição (case insensitive)
    palavras_excluir = [
        "COMO ENVIAR", "REVISÃO DE QUESTÕES", "CRACHÁ"
    ]

    # Função de filtro
    def filtrar_item(item):
        desc = item["descricao"].upper()

        if not any(p in desc for p in palavras_incluir):
            return False
        if any(p in desc for p in palavras_excluir):
            return False
        return True

    # Extração dos dados
    resultados = []
    for item in itens:
        if filtrar_item(item):
            hash_arquivo = item["nomeArquivoHash"]
            descricao = item['descricao']
            # Prioriza a extensão do campo tipo, se houver, senão pega do titulo
            extensao = item.get("tipo")
            if not extensao:
                extensao = re.search(r'\.([a-z0-9]+)$', item["titulo"], re.I)
                extensao = extensao.group(1) if extensao else ""
            resultados.append({'nome_arquivo':f'{hash_arquivo}.{extensao}','descricao':descricao})
    
    return resultados
```
