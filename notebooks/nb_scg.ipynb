```python
def limpar_duplicados(dicionario, chave_unica, valores_vistos=None):
    if valores_vistos is None:
        valores_vistos = set()

    chaves_para_remover = []
    for chave, valor in dicionario.items():
        if isinstance(valor, dict):
            # Recursivamente limpa subdicionários
            limpar_duplicados(valor, chave_unica, valores_vistos)
        elif chave == chave_unica:
            if valor in valores_vistos:
                # Marca o dicionário inteiro para remoção se o valor já foi visto
                return True
            valores_vistos.add(valor)

    # Se identificado para remoção, sinaliza ao chamador
    return False

def remover_duplicados_estrutura(data, chave_unica):
    if isinstance(data, dict):
        if limpar_duplicados(data, chave_unica):
            return None  # Retorna None se o item inteiro deve ser removido
        return data
    elif isinstance(data, list):
        nova_lista = []
        for item in data:
            resultado = remover_duplicados_estrutura(item, chave_unica)
            if resultado is not None:
                nova_lista.append(resultado)
        return nova_lista

# Exemplo de uso
dados = data["GERAL"]["SEMANA DE CONHECIMENTOS GERAIS"]
dados_limpados = remover_duplicados_estrutura(dados, "ENUNCIADO")
path_file_limpos = os.path.abspath(os.getcwd())+"\\BD\\atividades\\202451\\teste2.json"
save_json(data=dados_limpados,path_file=path_file_limpos)
```
